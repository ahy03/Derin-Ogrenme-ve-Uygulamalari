## Neden Tekrarlayan Sinir Ağları?
RNN, ileri beslemeli sinir ağında birkaç sorun olduğu için oluşturuldu:

- Sıralı verileri işleyemez
- Yalnızca geçerli girişi dikkate alır
- Önceki girişler ezberlenemiyor
- Bu sorunların çözümü RNN'dir. 
    - Bir RNN, mevcut girdi verilerini ve önceden alınan girdileri kabul ederek sıralı verileri işleyebilir. 
    - RNN'ler, dahili bellekleri nedeniyle önceki girişleri ezberleyebilir.

## Dizi(sıra) Tahmin Problemleri (Sequence Prediction Problems)

- Dizi tahmini, daha önce gözlemlenen sembol sırasına dayalı olarak sonraki sembol(ler)i tahmin etmekten oluşan popüler bir makine öğrenimi görevidir. Bu semboller bir sayı, alfabe, kelime, olay veya web sayfası veya ürün gibi bir nesne olabilir. 
- Örneğin:
    - Bir metindeki kelime veya karakter dizisi
    - Bir müşteri tarafından satın alınan bir dizi ürün
    - Günlüklerde gözlemlenen bir dizi olay

- Dizi tahmini, diğer denetimli öğrenme problemlerinden farklıdır, çünkü modelleri eğitirken ve tahminlerde bulunurken verilerdeki sıranın korunmasını zorunlu kılar.
- Dizi tahmini, çeşitli endüstrilerde gerçek yaşam uygulamaları bulan yaygın bir sorundur. 
- Bu kısımda üç tür dizi tahmini problemini konuşacağız:
    1. Bir sonraki değeri tahmin etme
    2. Bir sınıf etiketini tahmin etme
    3. Bir diziyi tahmin etme

### Bir sonraki değeri tahmin etme
- Bir dizinin sonraki öğesini tahmin edebilmek birçok uygulamada önemli bir sorudur. 
- Bir dizi tahmin modeli, sıralı girdi verilerindeki modeli tanımlamayı ve bir sonraki değeri tahmin etmeyi öğrenir.

### Zaman serisi tahmini
- Zaman serisi, gözlem dizisinin zaman boyutunda sırayla olduğu sıralı bir veri serisini ifade eder. 
- Zaman serisi tahmini, seride bir sonraki adımda ne olacağına dair tahminler yapmakla ilgilidir. 
- Bu nedenle, Zaman serisi tahmini, modeli geçmişte zamana bağlı veriler üzerinde eğitmeyi ve bunları gelecekteki gözlemleri tahmin etmek için kullanmayı içerir.
- Örneğin, belirli bir aydaki hisse senedi fiyatlarını tahmin etmek.

- **Ancak**, zaman serisi tahminini bir regresyon probleminden farklı kılan iki öğe vardır.
1. Zaman serisi zamana bağlıdır ve zamana göre sıralanır. 
    - Ancak, hedef değişkenin özellikler tarafından alınan değerlere bağlı olduğu sıralı olmayan verilere Regresyon uygulanabilir.
2. Zaman serileri mevsimsellik trendlerini arar. 
    - Örneğin, gündüz elektrik talebi geceleri düşecek ve yaz aylarında hava yolcu sayısı artacaktır.

### Standart RNN'lerin İki Sorunu
1. Kaybolan Gradyan Problemi
    - Tekrarlayan Sinir Ağları, borsa tahmini, makine çevirisi ve metin oluşturma gibi zamana bağlı ve sıralı veri sorunlarını modellemenizi sağlar. 
    - Bununla birlikte, eğim sorunu nedeniyle RNN'nin eğitilmesinin zordur.
    - RNN'ler, kaybolan gradyanlar probleminden muzdariptir. 
        - Gradyanlar, RNN'de kullanılan bilgileri taşır ve gradyan çok küçük olduğunda parametre güncellemeleri önemsiz hale gelir. Bu, uzun veri dizilerinin öğrenilmesini zorlaştırır.     
\begin{equation}
w_{n+1}=w_{n}-\eta \frac{df(x)}{dx}
\end{equation}

Daha genel olarak,
\begin{equation}
w_{n+1}=w_{n}-\eta \nabla f(w_{n})
\end{equation}
burada
\begin{equation}
\nabla f(w_{n}(x,y))=\begin{bmatrix}\frac{\partial f(w_{n})}{\partial x} \\ \frac{\partial f(w_{n})}{\partial y} \end{bmatrix}
\end{equation}.

2. Patlayan Gradyan Problemi

Bir sinir ağını eğitirken, eğim azalmak yerine üssel olarak büyüme eğilimindeyse buna Patlayan Gradyan denir. Bu sorun, büyük hata gradyanları biriktiğinde ortaya çıkar ve eğitim süreci sırasında sinir ağı modeli ağırlıklarında çok büyük güncellemelere neden olur.

Gradyan problemlerinde uzun eğitim süresi, düşük performans ve kötü doğruluk başlıca sorunlardır.

#### Uzun Süreli Bağımlılıklar.
- Metindeki son kelimeyi tahmin etmek istediğinizi varsayalım: 
    - “Bulutlar ______ içindedir.”
- Bunun en bariz cevabı “**gökyüzü**”dür. Yukarıdaki cümledeki son kelimeyi tahmin etmek için başka bir bağlama ihtiyacımız yok.
- Şimdi ise şu cümleyi düşünün: 
    - “Son 10 yıldır İspanya'da kalıyorum. ..... Akıcı ______ konuşabiliyorum.”
- Tahmin edeceğimiz kelime, bağlamdaki önceki birkaç kelimeye bağlı olacaktır. Burada metindeki son sözü tahmin etmek için İspanya bağlamına ihtiyacınız var ve bu cümleye en uygun cevap “İspanyolca”. İlgili bilgi ile ihtiyaç duyulan nokta arasındaki boşluk çok büyümüş olabilir. **LSTM**'ler bu sorunu çözmenize yardımcı olur.

```python
import pandas as pd

import numpy as np
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import math
import matplotlib.pyplot as plt
```

```python
def create_RNN(hidden_units, dense_units, input_shape, activation):
    model = Sequential()
    model.add(SimpleRNN(hidden_units, input_shape=input_shape, 
                        activation=activation[0]))
    model.add(Dense(units=dense_units, activation=activation[1]))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
```

```python
demo_model = create_RNN(2, 1, (3,1), activation=['linear', 'linear'])
```

```python
demo_model.summary()
```

```python
wx = demo_model.get_weights()[0]
wh = demo_model.get_weights()[1]
bh = demo_model.get_weights()[2]
wy = demo_model.get_weights()[3]
by = demo_model.get_weights()[4]
  
print('wx = ', wx, ' wh = ', wh, ' bh = ', bh, ' wy =', wy, 'by = ', by)
```

```python
x = np.array([1, 2, 3])
# Reshape the input to the required sample_size x time_steps x features 
x_input = np.reshape(x,(1, 3, 1))
y_pred_model = demo_model.predict(x_input)
```

```python
y_pred_model
```

```python
 
m = 2
h0 = np.zeros(m)
h1 = np.dot(x[0], wx) + h0 + bh
h2 = np.dot(x[1], wx) + np.dot(h1,wh) + bh
h3 = np.dot(x[2], wx) + np.dot(h2,wh) + bh
o3 = np.dot(h3, wy) + by
 
print('h1 = ', h1,'h2 = ', h2,'h3 = ', h3)
 
print("Prediction from network ", y_pred_model)
print("Prediction from our computation ", o3)
```

```python
# Parameter split_percent defines the ratio of training examples
def get_train_test(url, split_percent=0.8):
    df = pd.read_csv(url, usecols=[1], engine='python')
    data = np.array(df.values.astype('float32'))
    scaler = MinMaxScaler(feature_range=(0, 1))
    data = scaler.fit_transform(data).flatten()
    n = len(data)
    # Point for splitting data into train and test
    split = int(n*split_percent)
    train_data = data[range(split)]
    test_data = data[split:]
    return train_data, test_data, data
```

```python
sunspots_url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/monthly-sunspots.csv'
```

```python
train_data, test_data, data = get_train_test(sunspots_url)
```

```python
# Prepare the input X and target Y
def get_XY(dat, time_steps):
    # Indices of target array
    Y_ind = np.arange(time_steps, len(dat), time_steps)
    Y = dat[Y_ind]
    # Prepare X
    rows_x = len(Y)
    X = dat[range(time_steps*rows_x)]
    X = np.reshape(X, (rows_x, time_steps, 1))    
    return X, Y

time_steps = 12
trainX, trainY = get_XY(train_data, time_steps)
testX, testY = get_XY(test_data, time_steps)
```

```python
model = create_RNN(hidden_units=3, dense_units=1, input_shape=(time_steps,1), 
                   activation=['tanh', 'tanh'])
```

```python
model.summary()
```

```python
model.fit(trainX, trainY, epochs=20, batch_size=1, verbose=2)
```

```python
def print_error(trainY, testY, train_predict, test_predict):    
    # Error of predictions
    train_rmse = math.sqrt(mean_squared_error(trainY, train_predict))
    test_rmse = math.sqrt(mean_squared_error(testY, test_predict))
    # Print RMSE
    print('Train RMSE: %.3f RMSE' % (train_rmse))
    print('Test RMSE: %.3f RMSE' % (test_rmse))    
```

```python
# make predictions
train_predict = model.predict(trainX)
test_predict = model.predict(testX)
# Mean square error
print_error(trainY, testY, train_predict, test_predict)
```

```python
# Plot the result
def plot_result(trainY, testY, train_predict, test_predict):
    actual = np.append(trainY, testY)
    predictions = np.append(train_predict, test_predict)
    rows = len(actual)
    plt.figure(figsize=(15, 6), dpi=80)
    plt.plot(range(rows), actual)
    plt.plot(range(rows), predictions)
    plt.axvline(x=len(trainY), color='r')
    plt.legend(['Actual', 'Predictions'])
    plt.xlabel('Observation number after given time steps')
    plt.ylabel('Sunspots scaled')
    plt.title('Actual and Predicted Values. The Red Line Separates The Training And Test Examples')
plot_result(trainY, testY, train_predict, test_predict)
```
