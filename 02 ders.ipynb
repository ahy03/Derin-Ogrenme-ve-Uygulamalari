# Derin Öğrenme Kavramı 

- Katmanlar: Ağı oluşturan yapılar
- Girdi verisi ve hedefler
- Kayıp Fonksiyonu: Öğrenmeyi sağlayan geribildirim
- Eniyileme algoritması: Öğrenmenin nasıl gerçekleşeceğini belirler


![ ](imgs/loss.PNG)

- Ağ, zincirleme eklenen katmanlardan oluşur ve girdi verisini tahminlere eşler. 
- Kayıp fonksiyon yardımıyla tahminler ile hedefler arasındaki fark olan kayıp değeri hesaplanarak ağın tahminlerinin hedefleri ne kadar karşıladığı ölçülür. 
- Eniyileme algoritması kayıp değerini kullanarak ağırlıkları günceller.

**Katman:** Bir veya birden fazla tensörü girdi olarak alarak sonuçta bir veya daha fazla katmana dönüştüren veri işleyen modüldür.

![ ](imgs/nn.JPG)

- Sinir ağının giriş ile çıktı katmanı arasında kalan katmanlar gizli katmanlardır. 
- Gizli katmanların sayısı modelin derinliğini verir. 
- Derin sinir ağındaki derinlik kavramı buradan gelmektedir. 
- Derin sinir ağının son katmanı çıktı katmanıdır


### Tensör

- Bazı katmanlar durum bilgisi taşımaz (stateless). 
- Ancak çoğu katmanda stokastik gradyan inişi sonucunda öğrenilen ve durum bilgisi olan bir veya birden çok tensörden oluşan ağırlıkları vardır.  
- Bu ağırlıklar ağın öğrendiği bilgiyi bulundurur.
- Farklı katmanlar farklı tensör biçimleri için uygundur.
 - 2BTensör (samples,features) *Tamamen Bağlı olan Dense Katmanında*
 - 3BTensör (samples,timesteps,features) *Yinelenen olan Recurrent*
 - 4BTensör (Görüntü verisi) *2BEvrişim olan Conv2D*

- Katmanla derin öğrenmenin Lego parçaları gibidir. 
- Her katmanın belli bir şekle sahip tensör girdisi ve belli şekle sahip çıktısı vardır.

- Katman dinamik olarak önceki katmanın çıktı boyutuna bakarak yeni katmanı hazırlayabilir.

```python
model.fit(X, y, batch_size=1, epochs=500) ##GPU veya CPU
print(model.predict(X))
```

- İkili bir sınıflandırma problemidir (diyabetin başlangıcının 1 olması veya 0 olmaması). 
- Her hastayı tanımlayan tüm girdi değişkenleri sayısaldır. 

- Örneklem sayısı: 768

- Özellik Sayısı: 8 
   - Girdi
       1. Hamile kalma sayısı
       2. Glikoz
       3. Kan basıncı (mm Hg)
       4. Deri kalınlığı (mm)
       5. İnsülin (mu U/ml)
       6. Vücut kitle indeksi (kg cinsinden ağırlık/(m cinsinden boy)^2)
       7. Soyumuzdaki kişilere göre diyabet olma ihtimalimizi hesaplayan bir fonksiyon
       8. Yaş (yıl)
   - Çıktı
       9. Sınıf (0 or 1)

```python
# Kütüphaneler yükleme
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split #scikit-learn
```

```python
# Veri seti yükleme
dataset =  pd.read_csv('imgs/diabetes.csv', delimiter=',')
```

```python
dataset.head(5)
```

```python
# Girdi ve çıktı parametrelerini belirleme
X = dataset.iloc[:,0:8]
y = dataset.iloc[:,8]
```

```python
#Eğitim-test verisi ayırma
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.3)
```

```python
# Model oluşturma
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu')) # Girdi ve 1. Gizli katman
model.add(Dense(8, activation='relu'))  #2. Gizli katman
model.add(Dense(1, activation='sigmoid')) #Çıktı katmanı
model.summary()
```

- Doğru yapılması gereken ilk şey, girdi katmanının doğru sayıda girdi özelliğine sahip olduğundan emin olmaktır. 
- input_dim argümanıyla ilk katman oluşturulurken ve 8 girdi değişkeni için 8 olarak ayarlanırken belirtilebilir.
- Bu örnekte, üç katmanlı tam bağlantılı bir ağ yapısı kullandık.
- Eskiden tüm katmanlar için Sigmoid ve Tanh aktivasyon fonksiyonları tercih ediliyordu. 
- Bu günlerde, ReLU etkinleştirme işlevi kullanılarak daha iyi performans elde edilmektedir. 
- Ağ çıktımızın 0 ile 1 arasında olduğundan ve sınıf 1 olasılığıyla eşleştirilmesinin kolay olduğundan çıktı katmanında bir sigmoid kullanırız.

```python
# Derleme
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Eğitme
history=model.fit(X_train, y_train, epochs=150, batch_size=10, validation_data = (X_val, y_val) ) #verbose=0
```

```python
plt.plot(history.history['accuracy'], label='Training accuracy')
plt.plot(history.history['val_accuracy'], label='Validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy-Epoch Graph')
plt.legend()
```

```python
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss-Epoch Graph')
plt.legend()
```

- Epoch : Eğitim veri kümesindeki tüm satırlardan bir geçiş.
- Batch: Ağırlıklar güncellenmeden önce model tarafından bir dönem içinde dikkate alınan bir veya daha fazla örnek.
- Eğitim süreci, epochs bağımsız değişkenini kullanarak belirtmemiz gereken, epochs adı verilen veri kümesi aracılığıyla sabit sayıda yineleme için çalışacaktır. 
- Ayrıca, her çağda model ağırlıkları güncellenmeden önce dikkate alınan veri kümesi satırlarının sayısını, toplu iş boyutu adı verilen ve batch_size bağımsız değişkenini kullanarak ayarlamalıyız.
- CPU veya GPU'nuzda işin gerçekleştiği yer burasıdır.

```python
# Değerlendirme
loss, accuracy = model.evaluate(X_val, y_val)
print('Accuracy: %.2f' % (accuracy*100))
#print('Loss: %.2f' % (loss*100))
```

- İdeal olarak, modelinizi eğitmek ve değerlendirmek için verilerinizi eğitim ve test veri kümelerine ayırabilirsiniz.
- evaluate() iki değer içeren bir liste döndürür. Birincisi, veri kümesindeki modelin kaybı, ikincisi ise veri kümesindeki modelin doğruluğu olacaktır. 
- Yalnızca doğruluğu bildirmekle ilgileniyoruz, bu nedenle kayıp değerini göz ardı edeceğiz.
- İdeal olarak, kaybın sıfıra ve doğruluğun 1.0'a (örneğin %100) gitmesini isteriz.
- Modelimi eğittikten sonra, onu yeni veriler üzerinde tahminler yapmak için nasıl kullanabilirim?

```python
# Tahmin
predictions = (model.predict(X_val) > 0.5).astype(int)
```
